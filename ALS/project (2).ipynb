```python
import pandas as pd

# Load data from files
data_columns = ["user_id", "movie_id", "rating", "timestamp"]
data = pd.read_csv("data", sep="\t", names=data_columns)

genre_columns = ["genre_name", "genre_id"]
genre = pd.read_csv("genre", sep="|", names=genre_columns)

item_columns = ["movie_id", "title", "release_date", "video_release_date", "IMDb_URL", 
                "unknown", "Action", "Adventure", "Animation", "Children's", "Comedy", 
                "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", 
                "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"]
item = pd.read_csv("item", sep="|", names=item_columns, encoding="latin-1")
```

```python
# Assuming the genre columns start from index 5 (column "unknown") in the "item" DataFrame
genre_columns = item.columns[5:]

# Add a new column "genre" to the "item" DataFrame
item["genre"] = item.apply(lambda row: ",".join(genre for genre, value in zip(genre_columns, row[5:]) if value == 1), axis=1)

# Display the modified "item" DataFrame
print(item.head())
```

```python
item = item.drop(["video_release_date", "unknown", "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror","Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"], axis=1)
```

```python
# Merge "merged_data" with "item" on the "movie_id" column
final_data = pd.merge(data, item[["movie_id", "title", "release_date", "IMDb_URL", "genre"]], on="movie_id")

# Display the final merged DataFrame
print(final_data.head())
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.sql.functions import col

# Create a Spark session
spark = SparkSession.builder.appName("MovieRecommendation").getOrCreate()

# Convert the final_data DataFrame to a Spark DataFrame
spark_final_data = spark.createDataFrame(final_data)

# Split the data into training and testing sets
(training, test) = spark_final_data.randomSplit([0.8, 0.2])

# Build the recommendation model using ALS on the training data
als = ALS(maxIter=10, regParam=0.01, userCol="user_id", itemCol="movie_id", ratingCol="rating", coldStartStrategy="drop")
model = als.fit(training)

# Evaluate the model by computing the Root Mean Squared Error (RMSE) on the test data
predictions = model.transform(test)
evaluator = RegressionEvaluator(metricName="rmse", labelCol="rating", predictionCol="prediction")
rmse = evaluator.evaluate(predictions)
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Generate top movie recommendations for all users
userRecs = model.recommendForAllUsers(10)

# Show the top recommendations for the first user
userRecs.select("user_id", "recommendations.movie_id").show(truncate=False)
```

```python
def get_recommendations_and_titles(model, input_movie_title, spark_data, num_recommendations=10):
    # Get movie ID for the input title
    input_movie_id = get_movie_id(input_movie_title, spark_data)
    
    if input_movie_id is not None:
        # Generate recommendations for the specified movie ID
        movie_rec = model.recommendForItemSubset(spark.createDataFrame([(input_movie_id,)], ["movie_id"]), num_recommendations)
        
        # Extract the recommended movie IDs
        recommended_movie_ids = [row[0] for row in movie_rec.collect()[0]['recommendations']]
        
        # Get unique movie titles for the recommended movie IDs
        recommended_movie_titles = spark_data.filter(spark_data['movie_id'].isin(recommended_movie_ids)).select('movie_id', 'title').distinct().collect()
        
        return recommended_movie_ids, recommended_movie_titles

# Example: Get recommendations and titles for a given movie title
input_movie_title = input('movie name:')
recommendations, recommended_movie_titles = get_recommendations_and_titles(model, input_movie_title, spark_final_data)

if recommendations is not None:
    print("\nRecommended Movie Titles:")
    for row in recommended_movie_titles:
        print(f"Title: {row['title']}")
```
